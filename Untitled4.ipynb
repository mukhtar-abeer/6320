```python
import numpy as np

# Decision values and probabilities
decision_values = [178015.5, 177619]
probabilities = [0.5, 0.5]

# Reference point
R = 100000

# Calculate Expected Utility (EU)
EU = 0.5 * (1 - np.exp(-decision_values[0] / R)) + 0.5 * (1 - np.exp(-decision_values[1] / R))

print(f'Expected Utility (EU): {EU:.2f}')
```

```python
# Calculate Certainty Equivalent (CE)
CE = -R * np.log(1 - EU)

# Calculate Risk Premium (RP) - Assuming decision 1 is chosen
EV = decision_values[0]  # Expected value for decision 1
RP = CE - EV

print(f'Certainty Equivalent (CE): {CE:.2f}')
print(f'Risk Premium (RP): {RP:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Utility function
def utility(x, R):
    return 1 - np.exp(-x/R)

# Decision values and probabilities
decision_values = [178015.5, 177619]
probabilities = [0.5, 0.5]

# Reference point
R = 100000

# Expected Utility (EU) calculation
EU = 0.5 * (1 - np.exp(-decision_values[0] / R)) + 0.5 * (1 - np.exp(-decision_values[1] / R))

# Certainty Equivalent (CE) calculation
CE = -R * np.log(1 - EU)

# Risk Premium (RP) calculation
EV = decision_values[0]  # Expected value for decision 1
RP = CE - EV

# Plotting
x_values = np.linspace(0, 2 * decision_values[0], 1000)
y_values = utility(x_values, R)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Risk Utility Curve')
plt.axhline(y=EU, color='r', linestyle='--', label='Expected Utility (EU)')
plt.axvline(x=CE, color='g', linestyle='--', label='Certainty Equivalent (CE)')
plt.plot([CE, EV], [EU - 0.05, EU - 0.05], color='orange', linestyle='--') # Risk Premium line
plt.scatter([CE], [EU], color='g', label='Certainty Equivalent (CE)')
plt.scatter([EV], [EU], color='r', label='Expected Value (EV)')
plt.scatter([CE], [EU - 0.05], color='orange', label='Risk Premium (RP)')
plt.xlabel('Wealth (x)')
plt.ylabel('Utility')
plt.title('Risk Utility Curve')
plt.legend()
plt.grid(True)

# Annotation for Risk Attitude
if RP > 0:
    risk_attitude = 'Risk-Seeking'
elif RP < 0:
    risk_attitude = 'Risk-Averse'
else:
    risk_attitude = 'Risk-Neutral'

plt.text(CE + 500, EU, f'CE ({CE:.2f})', fontsize=10, va='bottom', ha='left', color='blue')
plt.text(CE + 500, EU - 0.05, f'RP ({RP:.2f})', fontsize=10, va='top', ha='left', color='orange')
plt.text(CE + 500, EU - 0.1, risk_attitude, fontsize=10, va='top', ha='left', color='purple')

plt.show()

print(f'Expected Utility (EU): {EU:.2f}')
print(f'Certainty Equivalent (CE): {CE:.2f}')
print(f'Risk Premium (RP): {RP:.2f}')
print(f'Risk Attitude: {risk_attitude}')

```
