```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equations for EV(d_1) and EV(d_2)
def EV_d1(p):
    return (p * 142352) + ((1 - p) * 142552)

def EV_d2(p):
    return (p * 142622) + ((1 - p) * 142412)

# Generate values for p
p_values = np.linspace(0, 1, 1000)

# Calculate EV(d_1) and EV(d_2) for each value of p
EV_d1_values = EV_d1(p_values)
EV_d2_values = EV_d2(p_values)

# Find the intersection point
intersection_index = np.argmin(np.abs(EV_d1_values - EV_d2_values))

# Plot the sensitivity curve
plt.figure(figsize=(10, 6))
plt.plot(p_values, EV_d1_values, label='EV(d_1)')
plt.plot(p_values, EV_d2_values, label='EV(d_2)')
plt.scatter(p_values[intersection_index], EV_d1_values[intersection_index], color='red', label='Intersection Point')
plt.xlabel('Probability (p)')
plt.ylabel('Expected Value (EV)')
plt.title('Sensitivity Curve')
plt.legend()
plt.grid(True)
plt.show()

# Print the value of p at the intersection point
intersection_point_p = p_values[intersection_index]
print(f"The value of p at the intersection point is approximately: {intersection_point_p:.3f}")
```
