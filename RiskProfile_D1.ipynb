```python
import matplotlib.pyplot as plt

# Define the points
points = [(142351.6, 0.3), (142551.6, 0.7)]

# Separate x and y coordinates
x = [point[0] for point in points]
y = [point[1] for point in points]

# Plot the points
plt.scatter(x, y, color='blue')
plt.xlabel('Profit (million USD)')
plt.ylabel('Probability')
plt.title('Risk Profile')

# Draw lines from x-axis to points
for point in points:
    plt.hlines(point[1], 0, point[0], linestyle='dotted', colors='gray')
    plt.vlines(point[0], 0, point[1], linestyle='dotted', colors='gray')

# Set x-axis limit
plt.xlim(142300, 142600)

# Show grid
plt.grid(True)

# Show plot
plt.show()
```
